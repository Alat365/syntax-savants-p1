```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read CSV files.
beer_file= ("Resources/beers.csv")
brewery= ("Resources/breweries.csv")
beer_file= pd.read_csv(beer_file)
brewery_file= pd.read_csv(brewery)
```

```python
# Show the data from the file
beer_file.head()
```

```python
#ABV and bitterness affect on sale amount
df_beer_file = pd.read_csv("Resources/beers.csv")
df_beer_file.head()
```

```python
beer_file_df = beer_file.loc[beer_file['abv'] != "?"]
beer_file_df.head()
```

```python
beer_file_df = beer_file.set_index('abv')

del beer_file_df['brewery_id']

beer_file_df.head()
```

```python
beer_file.plot(kind = "scatter", x='abv', y='ibu', grid=True, figsize=(8,8), title='Beer vs IBU')
plt.show()
```

```python
beer_abv_df = pd.DataFrame(beer_file)
beer_abv_df
ipa_df = beer_abv_df.loc[beer_abv_df['style'] == "American IPA"]
ipa_df
```

```python
#needed to get the ipa data 
```

```python
beers = ipa_df["name"].head(15)
abv= ipa_df["abv"].head(15)
plt.bar(beers, abv, color='blue', align = 'center')
plt.xticks(rotation=90)
plt.xlabel("Beer Styles")
plt.ylabel("ABV")
plt.title("ABVs of American IPA")
plt.yticks([.02, .04, .06, .08])
plt.show()
```

```python
#needed to get the blonde ale data
```

```python
blonde_ale_df = ipa_df = beer_abv_df.loc[beer_abv_df['style'] == "American Blonde Ale"]
blonde_ale_df

beers = blonde_ale_df["name"].head(15)
abv= blonde_ale_df["abv"].head(15)
plt.bar(beers, abv, color='blue', align = 'center')
plt.xticks(rotation=90)
#labels
plt.xlabel("Beer Styles")
plt.ylabel("ABV")
plt.title("ABVs of American Blonde Ales")
plt.yticks([.02, .04, .06, .08])
plt.show()
```
