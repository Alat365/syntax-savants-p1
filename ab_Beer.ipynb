```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read CSV files.
beer_file= ("Resources/beers.csv")
brewery= ("Resources/breweries.csv")
beer_file= pd.read_csv(beer_file)
brewery_file= pd.read_csv(brewery)
```

```python
# Show the data from the file
beer_file.head()
```

```python
# Show the data from the file
brewery_file.head()
```

```python
#Let's talk types of beers. What are the different styles? Which are peoples favorite?

#display the types of beers

beer_styles = beer_file["style"].value_counts()
beer_styles

```

```python
#organize colums for dataframe

beer_styles_org= beer_styles.rename_axis("Beer Styles").reset_index(name="Total")
beer_styles_org
```

```python
#make it a dataframe
beer_styles_df = pd.DataFrame(beer_styles_org)
beer_styles_df

```

```python


#which are 10 of the most popoular beers?

beer_styles_df.head(10)

```

```python
#conclusion American IPAs are the most drank!

#what are ten of the least popular? Assuming that if there are more of these kinds created, that they are more widely popular.

beer_styles_df.tail(10)

#the least popular!
```

```python
#let's visualize our top beer styles!

beers = beer_styles_df["Beer Styles"].head(10)
total = beer_styles_df["Total"].head(10)
plt.bar(beers, total, color='green', align = 'center')
plt.xticks(rotation=90)


#labels

plt.xlabel("Beer Styles")
plt.ylabel("Totals")
plt.title("Top 10 Styles of Beer")

plt.show()
```

```python
top_ten_beers = beer_styles_df.head(10)

labels = top_ten_beers["Beer Styles"]
sizes = top_ten_beers["Total"]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.title("Percentage of Top 10 Beers Drank")
plt.show()
```
